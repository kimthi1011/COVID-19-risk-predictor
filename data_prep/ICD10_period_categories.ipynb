```python
import pandas as pd
import numpy as np
```

# positive data

```python
positive_condition_data = pd.read_csv(
    '/data/project/ubrite/covid19-hackathon/Team4_staging_area/parsed_data/ICD10_grouped_positive.csv'
).set_index('PERSON_ID')
```

```python
positive_condition_data.head()
```

```python
positive_condition_data["icd"] = positive_condition_data['CONCEPT_CODE'].str.split('.').str[0]
```

```python
positive_condition_data.head()
```

```python
positive_condition_data = positive_condition_data.drop(['CONCEPT_CODE','group'], axis=1)
```

```python
positive_condition_cat = pd.get_dummies(positive_condition_data).groupby(level=0).max()
```

```python
positive_condition_cat.head()
```

```python
positive_condition_cat.to_csv('/data/project/ubrite/covid19-hackathon/Team4_staging_area/parsed_data/positive_period_group_categorical.csv'
                      , index=True)
```
