```python
import pandas as pd
import numpy as np
```

# positive data

```python
positive_substance_data = pd.read_csv(
    '/data/project/ubrite/covid19-hackathon/Team4_staging_area/parsed_data/observation_positive.csv'
).set_index('PERSON_ID')
```

```python
positive_substance_data.head(10)
```

```python
# remove substance abuse
positive_substance_data = positive_substance_data[~positive_substance_data['OBSERVATION_SOURCE_VALUE'].str.contains('Substance')]
```

```python
positive_substance_data["OBSERVATION_SOURCE_VALUE"] = positive_substance_data["OBSERVATION_SOURCE_VALUE"].str.split('.').str[0]
positive_substance_data.head(10)
```

```python
positive_substance_data['VALUE_AS_STRING'][positive_substance_data['VALUE_AS_STRING'].str.contains('Current')]="Current"
positive_substance_data["observation_combined"] = positive_substance_data["OBSERVATION_SOURCE_VALUE"] + " " + positive_substance_data["VALUE_AS_STRING"]
```

```python
positive_substance_data.head(50)
```

```python
positive_substance_data = positive_substance_data.drop(["VALUE_AS_STRING", "OBSERVATION_SOURCE_VALUE"], axis=1)
```

```python
positive_substance_cat = pd.get_dummies(positive_substance_data).groupby(level=0).sum()
positive_substance_cat.head()
```

```python
positive_substance_cat.to_csv('/data/project/ubrite/covid19-hackathon/Team4_staging_area/parsed_data/positive_observation_categorical.csv'
                      , index=True)
```
